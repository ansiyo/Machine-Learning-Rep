```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def softmax(x):
    exp_x = np.exp(x - np.max(x))  # Stability improvement
    return exp_x / exp_x.sum(axis=0)

# Input range
x = np.linspace(-10, 10, 100)

# Apply activation functions
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)

# Plot the functions
plt.figure(figsize=(12, 8))

# Sigmoid
plt.subplot(2, 2, 1)
plt.plot(x, y_sigmoid, label='Sigmoid', color='blue')
plt.title('Sigmoid Activation')
plt.grid()
plt.legend()

# Tanh
plt.subplot(2, 2, 2)
plt.plot(x, y_tanh, label='Tanh', color='orange')
plt.title('Tanh Activation')
plt.grid()
plt.legend()

# ReLU
plt.subplot(2, 2, 3)
plt.plot(x, y_relu, label='ReLU', color='green')
plt.title('ReLU Activation')
plt.grid()
plt.legend()

# Softmax demonstration (single-dimensional input)
x_softmax = np.array([1.0, 2.0, 3.0, 4.0, 1.0])
y_softmax = softmax(x_softmax)

print("Input to softmax:", x_softmax)
print("Output of softmax:", y_softmax)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def softmax(x):
    exp_x = np.exp(x - np.max(x))  # Stability improvement
    return exp_x / exp_x.sum(axis=0)

# Input range for visualization
x = np.linspace(-5, 5, 100)

# Example inputs for Softmax
x_softmax_example = np.array([-1.0, 2.0, 0.5, 0.3, -0.5])

# Apply activation functions
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)
y_softmax_example = softmax(x_softmax_example)

# Plot activation functions
plt.figure(figsize=(12, 8))

# Sigmoid
plt.subplot(2, 2, 1)
plt.plot(x, y_sigmoid, label='Sigmoid', color='blue')
plt.title('Sigmoid Activation')
plt.xlabel('Input (x)')
plt.ylabel('Output')
plt.grid()
plt.legend()

# Tanh
plt.subplot(2, 2, 2)
plt.plot(x, y_tanh, label='Tanh', color='orange')
plt.title('Tanh Activation')
plt.xlabel('Input (x)')
plt.ylabel('Output')
plt.grid()
plt.legend()

# ReLU
plt.subplot(2, 2, 3)
plt.plot(x, y_relu, label='ReLU', color='green')
plt.title('ReLU Activation')
plt.xlabel('Input (x)')
plt.ylabel('Output')
plt.grid()
plt.legend()

# Softmax
plt.subplot(2, 2, 4)
plt.bar(range(len(x_softmax_example)), y_softmax_example, color='red', alpha=0.7)
plt.xticks(range(len(x_softmax_example)), labels=[f"Class {i+1}" for i in range(len(x_softmax_example))])
plt.title('Softmax Activation (Example)')
plt.ylabel('Probability')
plt.grid()

plt.tight_layout()
plt.show()

# Print Softmax example result
print("Input to Softmax:", x_softmax_example)
print("Output of Softmax (probabilities):", y_softmax_example)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a simple dataset
X = np.array([[0.5], [2.0], [3.0]])  # Inputs
y = np.array([[0.8], [0.3], [0.9]])  # Targets
# Inputs (): These are the input values for the neural network.Example:  = [0.5,2.0,3.0]
# Targets (y): These are the true values that the neural network tries to predict.Example:y=[0.8,0.3,0.9].

# Initialize weights and biases
weights = np.random.rand(1, 1)
bias = np.random.rand(1)

# Learning rate
learning_rate = 0.1

# Activation function: Sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of sigmoid
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))
# This is needed during backpropagation to compute how the loss changes with respect to weights and biases.

# Loss function: Mean Squared Error (MSE)
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
# Measures the average squared difference between the true values and predicted values.

# Training loop
epochs = 1000 # 1000 epochs (iterations)
losses = []  # Store losses for visualization

for epoch in range(epochs):
    # Forward pass
    z = np.dot(X, weights) + bias  # Linear combination
    predictions = sigmoid(z)       # Activation function
# Linear Combination: This is the weighted sum of the inputs plus the bias: z=Xw+b
# Predictions: The sigmoid activation function is applied to z to get predictions between 0 and 1.

    # Compute the loss
    loss = mse_loss(y, predictions)
    losses.append(loss)
# Compute how far off the predictions are from the true values using MSE.
# Save the loss to visualize later.

    # Backpropagation
    d_loss_d_pred = 2 * (predictions - y) / y.size  # Derivative of MSE w.r.t predictions
# Measures how much the loss changes when the predictions change.
    d_pred_d_z = sigmoid_derivative(z)             # Derivative of sigmoid
# Measures how much the predictions change with respect to Z.
    d_loss_d_z = d_loss_d_pred * d_pred_d_z        # Chain rule: dL/dz
# Combines the above gradients using the chain rule:/ = /predictions * predictions/z

    # Gradients for weights and bias
    d_loss_d_weights = np.dot(X.T, d_loss_d_z)  # Gradient for weights
    d_loss_d_bias = np.sum(d_loss_d_z, axis=0)  # Gradient for bias

    # Update weights and biases
    weights -= learning_rate * d_loss_d_weights
    bias -= learning_rate * d_loss_d_bias
# These measure how the loss changes with respect to the weights and bias.

# Final weights and bias
print("Final Weights:", weights)
print("Final Bias:", bias)

# Final predictions
final_predictions = sigmoid(np.dot(X, weights) + bias)
print("\nFinal Predictions:", final_predictions)

# Visualization
plt.figure(figsize=(12, 6))

# Loss curve
plt.subplot(1, 2, 1)
plt.plot(losses, color='blue')
plt.title("Loss Over Epochs")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.grid()
# Shows how the loss decreases over epochs, indicating learning.

# Predictions vs Targets
plt.subplot(1, 2, 2)
plt.scatter(range(len(y)), y, color='red', label='Targets')
plt.scatter(range(len(final_predictions)), final_predictions, color='green', label='Predictions')
plt.title("Targets vs Predictions")
plt.xlabel("Data Points")
plt.ylabel("Values")
plt.legend()
plt.grid()

plt.show()
```
