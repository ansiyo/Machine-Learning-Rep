```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named df
# If the 'date' column is not already a datetime object, convert it
df['year'] = pd.to_datetime(df['Date'])

# Extract the day of the year and year for grouping
df['day_of_year'] = df['Date'].dt.dayofyear
df['year'] = df['Date'].dt.year

# Plot for a selected variable, e.g., 'meantemp'
plt.figure(figsize=(12, 8))

for year in df['Date'].unique():
   year_data = df[df['Date'] == year]
   year_data = year_data.resample("1m").median()
   plt.plot(year_data['day_of_year'], year_data['Site No'], label=f'{year}')

plt.title("Distribution of e-waste dumping sites in Kenya")
plt.xlabel("Date")
plt.ylabel("Site No")
plt.legend()
plt.show()
```
